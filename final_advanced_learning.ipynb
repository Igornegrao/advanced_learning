DATA PROCESSING

```python
## Lets first handle with outliers.

data = df
```

```python
features_with_outliers = ['creatinine_phosphokinase','ejection_fraction','platelets','serum_creatinine','serum_sodium']
for feature in features_with_outliers:
    q1 = data[feature].quantile(0.25)
    q3 = data[feature].quantile(0.75)
    IQR = q3-q1
    lower_limit = q1 - (IQR*1.5)
    upper_limit = q3 + (IQR*1.5)
    data.loc[data[feature]<lower_limit,feature] = lower_limit
    data.loc[data[feature]>upper_limit,feature] = upper_limit
```

```python
num_col = ["age","creatinine_phosphokinase","ejection_fraction","platelets","serum_creatinine","serum_sodium", "time"]

for i in num_col:
   
    fig, axs = plt.subplots(1,2,figsize=(15, 3))

    sns.histplot(data[i],bins=20, kde=True,ax=axs[0]);
    sns.boxplot(data[i], ax = axs[1], color='#99befd', fliersize=1);
```

- We can see that most outliers were removed

```python
## Lets now split the data into train and test.

X = data.drop(['death_event'], axis = 1)
y = data['death_event']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

- Next steps should be on handling imbalanced data
- We will use 2 different ways of handling unbalanced data:
    - SMOTE
    - ADASYN

```python
smote = SMOTE(k_neighbors=8)
```

```python
X_train_smote, y_train_smote = smote.fit_resample(X_train.astype('float'), y_train)
```

```python
from collections import Counter
print('Before SMOTE : ', Counter(y_train))
print('After SMOTE :', Counter(y_train_smote))
```

ADASYN

```python
ada = ADASYN(n_neighbors=8)
```

```python
X_train_ada, y_train_ada = ada.fit_resample(X_train.astype('float'), y_train)
```

```python
from collections import Counter
print('Before ADASYN : ', Counter(y_train))
print('After ADASYN :', Counter(y_train_ada))
```

Modelling Classifiers
- Our plan here is to use the RandonForrestClassifier to classify the events of death for the three datas created - unbalanced, SMOTE and ADASYN.

Randon Forrest

```python
model_unbalanced = RandomForestClassifier(max_depth=150, random_state=30)
model_smote = RandomForestClassifier(max_depth=150, random_state=30)
model_ada = RandomForestClassifier(max_depth=150, random_state=30)
```

```python
## Fitting the models

model_unbalanced.fit(X_train, y_train)
model_smote.fit(X_train_smote, y_train_smote)
model_ada.fit(X_train_ada, y_train_ada)
```

Evaluation

```python
y_unbalanced_pred = model_unbalanced.predict(X_test)
y_smote_pred = model_smote.predict(X_test)
y_ada_pred = model_ada.predict(X_test)
```
